```python
import torch
```

```python
# scalar(0-D tensors)
# in fact, Pytorch does not have a special tensor with zero dimension
x = torch.rand(10)
x.size()
```

```python
# vector(1-D tensors)
# we can use a vector to store you weight in the last:
weight_data = torch.FloatTensor([70.1, 71.0, 71.5, 71.2, 71.1, 70.5, 70.8])
weight_data.size()
```

```python
# matrix(2-D tensor)
# most of structured data is represented in matrix,for example house price
# here house_data:1 column->size, 2 column->bedroom number, 3 column->house price
import numpy as np
house_data = np.array([[60, 2, 120],
                      [75, 2, 130],
                      [90, 3, 160],
                      [100, 3, 200]])
tensor_data = torch.from_numpy(house_data) # np.array -> tensor
tensor_data.size()
```

```python
# 3-D tensor
# add matrixes together, 3D tensors is represent data-like image
# for example, a image shape is   ,where first index denote height, the second denote weight,
# the third denote channel
import os
from PIL import Image
import matplotlib.pyplot as plt
root = 'E:/1Practice/project_pytorch/image data'
img_path = os.path.join(root, '0.jpg')
# print(img_data) # check dir 
img = Image.open(img_path)
numpy_data = np.array(img)
tensor_data = torch.from_numpy(numpy_data)
print(tensor_data.size())  # pytorch h*w*c

# display image
image_data = tensor_data.numpy()
print(image_data.shape)
plt.figure()
plt.imshow(image_data)
plt.show()
```

# sliceing tensors
## get a portion of tensor

```python
# select one channel from image
plt.imshow(tensor_data[:,:,0].numpy()) 
plt.show()
```

```python
# crop image
plt.imshow(tensor_data[0:100, 50:250,:].numpy())
plt.show()
```
